```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import sklearn as skl
import tensorflow as tf
```

```python
# Generate dummy dataset
X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=78)

# Creating a DataFrame with the dummy data
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"] = y

# Plotting the dummy data
df.plot.scatter(x="Feature 1", y="Feature 2", c="Target", colormap="winter")
plt.show()
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

They call it "Sequential()" because you're stacking the various layers in a sequence.

```python
# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=2))
```

```python
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

School notes dictate that "binary_crossentropy" and "adam" optimizer are out of scope of this class. We will be using these as our default settings.

In short, cross-entropy is measuring the differences between probabilities of 2 distributions, one of which is observed and the other is predicted. It is binary because it has only 2 outcomes in the output.

Neural networks hide the complexity of these calculations away from you.

The "adam" optimizer is the optimization algorithm for stochastic gradient descent.

```python
# Create a DataFrame containing training history
history_df = pd.DataFrame(fit_model.history)

# Increase the index by 1 to match the number of epochs
history_df.index += 1

# Plot the loss
history_df.plot(y="loss")
plt.show()
```

This is plotting the number of epoch iterations against loss.

##### What are loss?
In layman terms, these are the errors we are looking at. Some of the common loss functions are:
* Mean absolute error
* Mean squared error
* Categorical cross entropy (Out of scope)

In short, the algorithm iterates itself and try to reduce its errors as much as possible to get an optimal result.

```python
# Plot the accuracy
history_df.plot(y="accuracy")
plt.show()
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
